Crie uma função que identifique se duas ou mais palavras inseridas pelo usuário são anagramas.

```python
def anagrama_check():

    palavra_1 = input("Escreva uma palavra: ")
    palavra_2 = input("Escreva uma segunda palavra: ")

    if (sorted(palavra_1) == sorted(palavra_2)):
        print(f"As palavras {palavra_1} e {palavra_2} são anagramas")
    else:
        print("As palavras não são anagramas")

anagrama_check()
```

Crie uma função que retorne a contagem de consoantes e vogais de uma palavra imputada pelo usuário.

```python
def contagem_letras():

    palavra = input("Digite uma palavra: ")
    palavra_min = palavra.lower()
    vogais = 0
    consoantes = 0

    for letra in palavra_min:
        if letra in "a, e, i, o, u":
            vogais = vogais + 1
        else:
            consoantes = consoantes + 1

    print(f"A palavra {palavra} tem o total de {vogais} vogais e {consoantes} consoantes")

contagem_letras()
```

Para que serve o argumento Kwargs nas funções?

```python
# o argumento Kwargs é a abreviação de keyword arguments e é usado geralmente para passar um dicionário com várias chaves(keys)
# para uma determinada função 

def func_kwargs(**kwargs):
    print(f'tipo: {type(kwargs)} conteudo: {kwargs}')
    for key, value in kwargs.items():
        print(f'atributo: {key}, valor: {value}')

func_kwargs(nome='Caio', sobrenome = 'Bolognese', idade = 21)
```

Escreva um programa em Python para verificar se um alfabeto é uma vogal ou uma consoante.

```python
vogais = ['a', 'e', 'i', 'o', 'u']
def verifica_alfabeto():
    letra = input('Insira uma letra do alfabeto: ')
    if letra in vogais:
        print(f'{letra} é uma vogal')
    else:
        print(f'{letra} é uma consoante')
verifica_alfabeto()
```
